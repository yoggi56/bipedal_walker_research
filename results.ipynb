```python
import plotly.express as px
import pandas as pd
import csv
import plotly.graph_objects as go
```

BipedalWalker2d

Reward plots

```python
# load CSV-data

# PPO
t_ppo = []
r_ppo = []
with open('statistics/bipedWalker2d_vPPO_rewards.csv', 'r', newline='') as csvfile:
    rewardreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in rewardreader:
        r_ppo.append(float(row[0]))
        t_ppo.append(float(row[1]))
        if float(row[1]) > 1000000:
            break

# DDPG
t_ddpg = []
r_ddpg = []
with open('statistics/bipedWalker2d_vDDPG_rewards.csv', 'r', newline='') as csvfile:
    rewardreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in rewardreader:
        r_ddpg.append(float(row[2]))
        t_ddpg.append(float(row[1]))
        if float(row[1]) > 1000000:
            break

# TD3
t_td3 = []
r_td3 = []
with open('statistics/bipedWalker2d_vTD3_rewards.csv', 'r', newline='') as csvfile:
    rewardreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in rewardreader:
        r_td3.append(float(row[2]))
        t_td3.append(float(row[1]))
        if float(row[1]) > 1000000:
            break

# SAC
t_sac = []
r_sac = []
with open('statistics/bipedWalker2d_vSAC_rewards.csv', 'r', newline='') as csvfile:
    rewardreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in rewardreader:
        r_sac.append(float(row[2]))
        t_sac.append(float(row[1]))
        if float(row[1]) > 1000000:
            break

# SAC PMTG
t_sac_pmtg = []
r_sac_pmtg = []
with open('./models/PMTG_SAC-bipedalWalker2d-model/best_single/2022-10-13_15-40-13/rewards.csv', 'r', newline='') as csvfile:
    rewardreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in rewardreader:
        r_sac_pmtg.append(float(row[0]))
        t_sac_pmtg.append(float(row[1]))
        if float(row[1]) > 10000000:
            break
r_sac_pmtg[:] = r_sac_pmtg[8:]
t_sac_pmtg[:] = t_sac_pmtg[8:]
```

```python
# PPO Reward Plot
df_ppo = pd.DataFrame(dict(
    x = t_ppo,
    y = r_ppo
))

fig = px.line(df_ppo, x="x", y="y", title="PPO Reward Function")
fig.update_layout(title='PPO Reward Function', title_x=0.5,
                   xaxis_title='Steps',
                   yaxis_title='Reward') 
fig.show()
```

```python
# DDPG Reward Plot
df_ddpg = pd.DataFrame(dict(
    x = t_ddpg,
    y = r_ddpg
))

fig = px.line(df_ddpg, x="x", y="y")
fig.update_layout(title='DDPG Reward Function', title_x=0.5,
                   xaxis_title='Steps',
                   yaxis_title='Reward') 
fig.show()
```

```python
# TD3 Reward Plot
df_td3 = pd.DataFrame(dict(
    x = t_td3,
    y = r_td3
))

fig = px.line(df_td3, x="x", y="y")
fig.update_layout(title='TD3 Reward Function', title_x=0.5,
                   xaxis_title='Steps',
                   yaxis_title='Reward') 
fig.show()
```

```python
# SAC Reward Plot
df_sac = pd.DataFrame(dict(
    x = t_sac,
    y = r_sac
))

fig = px.line(df_sac, x="x", y="y")
fig.update_layout(title='SAC Reward Function', title_x=0.5,
                   xaxis_title='Steps',
                   yaxis_title='Reward') 
fig.show()
```

```python
# SAC Reward Plot
df_sac_pmtg = pd.DataFrame(dict(
    x = t_sac_pmtg,
    y = r_sac_pmtg
))

fig = px.line(df_sac_pmtg, x="x", y="y", title="SAC Reward Function")
fig.update_layout(title='PMTG SAC Reward Function', title_x=0.5,
                   xaxis_title='Steps',
                   yaxis_title='Reward') 
fig.show()
```

```python
# All Reward Plots

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_sac["x"], y=df_sac["y"],
                    mode='lines',
                    name='SAC'))
fig.add_trace(go.Scatter(x=df_td3["x"], y=df_td3["y"],
                    mode='lines',
                    name='TD3'))
fig.add_trace(go.Scatter(x=df_ddpg["x"], y=df_ddpg["y"],
                    mode='lines',
                    name='DDPG'))
fig.add_trace(go.Scatter(x=df_ppo["x"], y=df_ppo["y"],
                    mode='lines',
                    name='PPO'))

fig.update_layout(title='Rewards for BipedalWalker2d-v3', title_x=0.5,
                   xaxis_title='Steps',
                   yaxis_title='Reward',
                   autosize=False,
                   width=800,
                   height=500
                   ) 
fig.show()
```
